```python
import pandas as pd
import matplotlib.pyplot as plt

# 데이터 로드
file_path = '/Users/jeonghyeonjeong/Documents/bigpl_team/data/감성분석/combined_sentiment_results.json'  # JSON 파일 경로를 여기에 입력
df = pd.read_json(file_path)

# LABEL_0과 LABEL_1을 분리
label_0 = df[df['sentiment_label'] == 'LABEL_0']
label_1 = df[df['sentiment_label'] == 'LABEL_1']

# 산점도 그리기
plt.figure(figsize=(10, 6))

# LABEL_0은 빨간색으로, 작은 점 크기로 표시
plt.scatter(label_0.index, label_0['sentiment_score'], color='red', label='LABEL_0', s=1, alpha=0.5)

# LABEL_1은 파란색으로, 작은 점 크기로 표시
plt.scatter(label_1.index, label_1['sentiment_score'], color='blue', label='LABEL_1', s=1, alpha=0.5)

plt.title('Sentiment Score by Sentiment Label')
plt.xlabel('Index')
plt.ylabel('Sentiment Score')
plt.legend()
plt.show()

```

```python
import json
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Load your JSON data
with open('/Users/jeonghyeonjeong/Documents/bigpl_team/data/감성분석/combined_sentiment_results.json', 'r', encoding='utf-8') as file:
    data = json.load(file)

# LABEL_0에 해당하는 텍스트 필터링 및 특정 문자 제거
texts = []
for entry in data:
    if entry['sentiment_label'] == 'LABEL_0':
        text = entry['text']
        # "SEP", "BR", "CLS" 제거
        text = text.replace("SEP", "").replace("br", "").replace("CLS", "").replace("ㅋ", "").replace("그냥", "").replace("다", "").replace("진짜", "").replace("지금", "").replace("정말", "").replace("참", "").replace("또", "")
        texts.append(text)

# 텍스트들을 하나의 문자열로 결합
combined_text = " ".join(texts)

# Create a word cloud
wordcloud = WordCloud(font_path='/Users/jeonghyeonjeong/Downloads/나눔 글꼴/나눔고딕/NanumFontSetup_OTF_GOTHIC/NanumGothic.otf', width=800, height=800, background_color='white').generate(combined_text)

# 워드 클라우드 표시
plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```
