```python
# csv저장, 속도 업


import csv
import concurrent.futures
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup

chrome_driver_path = '/Applications/chromedriver-mac-x64/chromedriver'
url_template = 'https://search.dcinside.com/post/p/{}/sort/latest/q/.EA.B9.80.ED.8F.AC.EC.84.9C.EC.9A.B8.ED.8E.B8.EC.9E.85'

chrome_options = Options()
chrome_options.add_argument('--headless')

driver = webdriver.Chrome(service=ChromeService(chrome_driver_path), options=chrome_options)

csv_file_path = 'dcinside_data.csv'

def scrape_page(url):
    driver.get(url)
    driver.implicitly_wait(5)
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    results = soup.select('ul.sch_result_list > li')

    rows = []
    for result in results:
        title_element = result.select_one('a.tit_txt')
        title = title_element.text if title_element else 'No title found'
        post_link = title_element['href']
        rows.append((title, post_link))

    return rows

def scrape_post(args):
    title, post_link = args
    driver.get(post_link)
    driver.implicitly_wait(5)
    post_html = driver.page_source
    post_soup = BeautifulSoup(post_html, 'html.parser')
    content_element = post_soup.select_one('div.write_div')
    content = content_element.text.strip() if content_element else 'No content found'
    return title, content

with open(csv_file_path, 'w', encoding='utf-8', newline='') as csvfile:
    csv_writer = csv.writer(csvfile)
    csv_writer.writerow(['Title', 'Content'])

    page = 1
    while True:
        url = url_template.format(page)
        page += 1

        rows = scrape_page(url)

        with concurrent.futures.ThreadPoolExecutor() as executor:
            post_results = list(executor.map(scrape_post, rows))

        for title, content in post_results:
            csv_writer.writerow([title, content])

        if not rows or not driver.find_elements_by_css_selector('div.bottom_paging_box a.page_next'):
            break

driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
from selenium.common.exceptions import NoSuchElementException
import time

# Chrome 드라이버 경로 설정 (본인의 환경에 맞게 수정하세요)
chrome_driver_path = '/Applications/chromedriver-mac-x64/chromedriver'

# Chrome 옵션 설정
chrome_options = Options()
chrome_options.add_argument('--headless')  # 화면을 띄우지 않고 실행할 경우 주석 처리

# Selenium Chrome 드라이버 초기화
driver = webdriver.Chrome(service=ChromeService(chrome_driver_path), options=chrome_options)

# 웹페이지 접속
url = 'https://search.dcinside.com/post/sort/latest/q/.EA.B9.80.ED.8F.AC.EC.84.9C.EC.9A.B8.ED.8E.B8.EC.9E.85'
driver.get(url)

# 페이지가 완전히 로드될 때까지 대기 (최대 10초)
driver.implicitly_wait(10)

# 명시적 대기를 위한 WebDriverWait 객체 생성
wait = WebDriverWait(driver, 5)

while True:
    # 페이지 소스코드 가져오기
    html = driver.page_source

    # BeautifulSoup을 사용하여 파싱
    soup = BeautifulSoup(html, 'html.parser')

    # 필요한 데이터 추출
    results = soup.select('ul.sch_result_list > li')

    for result in results:
        post_link = result.select_one('a.tit_txt')['href']

        # Navigate to the post link
        driver.get(post_link)

        # Wait for the post page to load
        driver.implicitly_wait(5)

        try:
            # Get the content from div.write_div
            post_html = driver.page_source
            post_soup = BeautifulSoup(post_html, 'html.parser')

            # Use result.select for the title
            title = result.select_one('a.tit_txt').text
            content = post_soup.select_one('div.write_div').text.strip()

            print(f'Title: {title}')
            print(f'Content: {content}')
            print('---' * 30)
        except AttributeError as e:
            print(f"Error: {e}")

    try:
        # Move to the next page
        next_button = driver.find_element(By.CLASS_NAME, 'page_next')
        next_button.click()

        # 지연 추가 (필요에 따라 조절)
        time.sleep(2)

        # 명시적 대기 사용: 현재 페이지가 더 이상 유효하지 않을 때까지 대기
        wait.until(EC.staleness_of(driver.find_element(By.CLASS_NAME, 'tit_txt')))
    except NoSuchElementException:
        # If there is no next button, break out of the loop
        break

# 브라우저 닫기
driver.quit()
```
