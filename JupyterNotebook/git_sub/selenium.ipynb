```python
import csv

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# 웹드라이버 경로를 자신의 환경에 맞게 설정
driver_path = '/Users/jeonghyeonjeong/Documents/4-1/CAPSTONE/chromedriver-mac-x64/chromedriver'

# Selenium 웹드라이버 설정
options = webdriver.ChromeOptions()
options.add_argument('--headless')  # 브라우저 창을 띄우지 않고 실행하려면 주석 처리
options.add_argument('--disable-gpu')
options.add_argument('--window-size=1200x600')

# 웹드라이버 실행
driver = webdriver.Chrome(options=options)

# 페이지 이동 및 크롤링
url = 'https://search.daum.net/search?w=news&q=%EA%B9%80%ED%8F%AC%EC%84%9C%EC%9A%B8%ED%8E%B8%EC%9E%85&enc=utf8&cluster=y&cluster_page=1&DA=DNS'
driver.get(url)

def crawl_page(driver, csv_writer):
    # 기사 목록 가져오기
    articles = WebDriverWait(driver, 10).until(
        EC.presence_of_all_elements_located((By.CSS_SELECTOR, 'ul.c-list-basic li[data-docid]'))
    )

    # 각 기사 정보 출력 및 CSV 파일에 저장
    for article in articles:
        title = article.find_element(By.CSS_SELECTOR, 'strong.tit-g a').text
        media = article.find_element(By.CSS_SELECTOR, 'a.item-writer strong.tit_item').text
        date = article.find_element(By.CSS_SELECTOR, 'span.gem-subinfo span.txt_info').text
        content = article.find_element(By.CSS_SELECTOR, 'p.conts-desc').text

        # CSV 파일에 쓰기
        csv_writer.writerow([title, media, date, content])

    # 다음 페이지로 이동
    try:
        next_button = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.CSS_SELECTOR, 'button.btn_next'))
        )
        if not next_button.get_attribute('disabled'):
            next_button.click()
            return True  # 다음 페이지로 이동했음을 반환
        else:
            return False  # 다음 페이지로 이동할 수 없음을 반환
    except Exception as e:
        print(f"An error occurred while trying to click the next button: {e}")
        return False

# CSV 파일 열기
csv_filename = 'daum_news_crawled_data.csv'
with open(csv_filename, 'w', newline='', encoding='utf-8') as csv_file:
    # CSV 작성자 생성
    csv_writer = csv.writer(csv_file)

    # 첫 번째 페이지 크롤링
    crawl_page(driver, csv_writer)

    # 다음 페이지로 이동하면서 크롤링
    while crawl_page(driver, csv_writer):
        pass

# 웹드라이버 종료
driver.quit()
```
