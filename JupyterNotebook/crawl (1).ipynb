```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
main_url = 'https://search.seoul.co.kr/index.php?keyword=%EA%B9%80%ED%8F%AC%20%ED%8E%B8%EC%9E%85&pageNum=' # 분석하고자 하는 페이지 주소
titles_list = []

# 페이지 범위
for page_number in range(1, 8):
    url = f'{main_url}{page_number}'
    res = requests.get(url)

    soup = BeautifulSoup(res.content, 'lxml')
    news_titles = soup.select("a") # 제목 부분

    for title in news_titles:
        titles_list.append(title.get_text()) # 리스트로 묶음

# 제목 출력
for title in titles_list:
    print(title)
print(len(titles_list))

df=pd.DataFrame([titles_list])
df.to_csv('seoulnp.csv')
```

```python

import requests
import pandas as pd
from bs4 import BeautifulSoup
main_url = 'https://www.gimpo.go.kr/news/selectBbsNttList.do?key=9377&bbsNo=466&searchCtgry=&pageUnit=10&searchCnd=SJ&searchKrwd=%ED%8E%B8%EC%9E%85&integrDeptCode=&&&pageIndex='
titles_list = []

# Loop through pages 1 to 5
for page_number in range(1, 3):
    url = f'{main_url}{page_number}'
    res = requests.get(url)

    # Process the response regardless of the status code
    soup = BeautifulSoup(res.content, 'lxml')
    news_titles = soup.select("span.p-media__heading-title")

    for title in news_titles:
        titles_list.append(title.get_text())

# Print the titles
for title in titles_list:
    print(title)
print(len(titles_list))

df=pd.DataFrame([titles_list])
df.to_csv('gimponewspotal.csv')
```

```python

import requests
import pandas as pd
from bs4 import BeautifulSoup
#https://www.igimpo.com/news/articleList.html?page=5&total=590&box_idxno=&sc_area=A&view_type=sm&sc_word=%ED%8E%B8%EC%9E%85
#https://www.igimpo.com/news/articleList.html?page=1&total=590&box_idxno=&sc_area=A&view_type=sm&sc_word=%ED%8E%B8%EC%9E%85
main_url = 'https://www.igimpo.com/news/articleList.html?page='

titles_list = []

# Loop through pages 1 to 5
for page_number in range(1, 6):
    url = f'{main_url}{page_number}'
    res = requests.get(url)

    # Process the response regardless of the status code
    soup = BeautifulSoup(res.content, 'lxml')
    news_titles = soup.select("h4.titles")

    for title in news_titles:
        titles_list.append(title.get_text())

# Print the titles
for title in titles_list:
    print(title)
print(len(titles_list))

df=pd.DataFrame([titles_list])
df.to_csv('gimponews.csv')
```
