```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib import font_manager

# CSV 파일 읽기 (파일 경로는 실제 파일 경로로 변경해주세요)
file_path_csv = '/content/inti_op.csv'
dataframe = pd.read_csv(file_path_csv, error_bad_lines=False)

# 'your_column_name'은 실제 텍스트가 있는 열의 이름으로 변경해야 합니다.
text_data = ' '.join(dataframe['text'].astype(str).tolist())
font_path = "/content/NanumGothic.otf"

# 워드클라우드 객체 생성
wordcloud = WordCloud(width=1000, height=600, background_color="white", font_path=font_path).generate(text_data)

# 워드클라우드를 시각화하여 보여줌
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
