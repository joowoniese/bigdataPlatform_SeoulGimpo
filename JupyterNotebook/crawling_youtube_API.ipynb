```python
# 영상 정보 수집

# 비디오ID
video_id_list = [res['items'][i]['id']['videoId'] for i in range(50)]
# 제목
video_title_list = [res['items'][i]['snippet']['title'] for i in range(50)]
# 설명
video_des_list = [res['items'][i]['snippet']['description'] for i in range(50)]

df_video = pd.DataFrame({'video_id':video_id_list,
                         'video_title':video_title_list,
                         'video_des':video_des_list})
df_video
```

```python
# 영상 정보 파일 저장
df_video.to_csv(path+"youtube_videos.csv")
```

```python
# 영상 댓글 수집

api_obj = build('youtube', 'v3', developerKey=KEY)

df_comments = pd.DataFrame()

for video_id in df_video['video_id']:   #수집된 영상 ID로 댓글 수집
  comments = list()
  try:
    response = api_obj.commentThreads().list(part='snippet,replies', videoId=video_id, maxResults=100).execute()

    while response:
      for item in response['items']:
          comment = item['snippet']['topLevelComment']['snippet']
          comments.append([video_id, comment['textDisplay'], comment['authorDisplayName'], comment['publishedAt'], comment['likeCount']])

          if item['snippet']['totalReplyCount'] > 0:
              for reply_item in item['replies']['comments']:
                  reply = reply_item['snippet']
                  comments.append([video_id, reply['textDisplay'], reply['authorDisplayName'], reply['publishedAt'], reply['likeCount']])

      if 'nextPageToken' in response:
          response = api_obj.commentThreads().list(part='snippet,replies', videoId=video_id, pageToken=response['nextPageToken'], maxResults=100).execute()
      else:
          break

    df_comments = df_comments.append(comments)

  except:
    continue

df_comments
```

```python
df_comments.reset_index(drop=True, inplace=True)  #인덱스 재조정

# 수집한 댓글 파일 저장
df_comments.to_csv(path+"youtube_comments.csv", header=['video_id', 'comment', 'user_id', 'time', 'num_likes'])
```
